# Content Based Filtering

```python
#Content based filtering recommends the food to the user based on user's preferences. It filters componenets, learn profile and analyses content.
# The system contract the two items and recommend if the ingredients and dish types are identicial. For an instance, a user might like Chicken Burger which contains bread, salads, chicken and so on. A chicken sandwich is also made up of similar items so that user might like sandwich too. 
# For that, we will merge name of the item with its additive and dish type which will form a different column in dataframe which will be used for comparision later.

def nameIngredient(x): #creating a method to combine the required attributes
    dish_type = x['dish_type'].lower().split(' ')
    dish_type.extend(x['dish_name'].lower().split(' '))
    dish_type.extend(x['additives'].lower().split(' '))
    return "".join(sorted(set(dish_type)))
dataframe1['NameTypeIngredient'] = dataframe1.apply(nameIngredient,axis=1)
dataframe1.head()
```
